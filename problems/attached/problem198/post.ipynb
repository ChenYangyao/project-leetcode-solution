```python
class VarDist:
    def __init__(self, fname = "dist.dat"):
        data = np.loadtxt( fname, dtype=float )
        self._tag = {}
        for i in data:
            key = tuple([int(i[0]),str(i[1]),str(i[2])])
            self._tag.setdefault( key, [] ).append( i[-3:] )
        for k,d in self._tag.items():
            self._tag[ k ] = np.array( d, dtype=float )
    def get(self):
        return self._tag
```

```python
vard = VarDist()
```

```python
x,y,e = vard.get()[(0,"100.0","99.0")].T

_, ax = m_get_figure()

ax.errorbar(
    x,y,yerr = (100*e,100*e), fmt = 'none', capthick = 3.0, capsize = 5.0, elinewidth = 1.5,
    c = mycms[0],label = r"$\rm Flat\ [0,200] $")

m_set_scale(ax,'log','log')
m_set_label(ax,'Security Link Length','Maximal Robbed Fraction')
ax.legend(loc = 'upper right')

plt.tight_layout()
plt.savefig('linklength.jpg')
plt.show()
```

```python
class VarNcorr:
    def __init__(self, fname = "ncorr.dat"):
        data = np.loadtxt( fname, dtype=float )
        self._tag = {}
        for i in data:
            key = int(i[0])
            self._tag.setdefault( key, [] ).append( i[-3:] )
        for k,d in self._tag.items():
            self._tag[ k ] = np.array( d, dtype=float )
    def get(self):
        return self._tag
```

```python
varn = VarNcorr()
```

```python
_, ax = m_get_figure()

cnt = 0
for k,d in varn.get().items():
    if k == 100 or k == 10000:
        cnt += 0
    else:
        print(cnt)
        x,y,e = d[-6:].T
        ax.plot(x,y,c=mycms[cnt], lw = 2.5)
        cnt += 1


m_set_scale(ax,'log','log')
m_set_label(ax,'Security Link Length','Maximal Robbed Fraction')
txpos(ax,'Corr-Length: ',(.42,.93), mycms[0], 16)
txposs(ax, ["None", "200", "500", "1000","2000","5000"], (.75,.93), (0.,-0.08), np.array(mycms), fonts=18)

plt.tight_layout()
plt.savefig('corlength.jpg')
plt.show()
```
